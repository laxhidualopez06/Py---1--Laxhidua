Ejercicio de aproximación numérica de una raíz

```python
import numpy as np
```

```python
X = np.linspace(0,2, 1000)
X
```

```python
X.shape
```

```python
x_aprox = X[np.argmin(np.abs(f))] #np.argmin ---> Para encontrar el indice posición del valor minimo en un array
print(x_aprox)
```

Ejercicio 2. Error numérico en una aproximación

Objetivo: Calcular el error al aproximar raíz 2

```python
import math  # Proporciona funciones y constantes matemáticas avanzadas para trabajar con números reales
```

```python
real = math.sqrt(2)
aprox = 1.41

error_absoluto = abs(real - aprox)
error_relativo = error_absoluto / real
print(error_absoluto, error_relativo)
```

Ejercicio 3. Optimización numérica simple

Problema minimizar f(x)=(x-3)**2

```python
x = 0
alpha = 0.1

for _ in range(20):
  grad = 2 * (x - 3)
  x = x - alpha * grad
print(x)
```

Ejercicio 2: Precisión vs Exactitud

```python
import numpy as np
mediciones = np.array([9.8, 9.81, 9.79, 9.80])
valor_real = 9.81
media = np.mean(mediciones)
error = abs(media - valor_real)
print("Media: ", media)
print("Error absoluto: ", error)
```

Ejercicio 3: Sesgo numérico en datos

```python
import numpy as np
datos_reales = np.random.normal(0,1,1000)
sesgo = 0.5

print(datos_reales)

datos_sesgados = datos_reales + sesgo

print(datos_sesgados)

print("Media real:", np.mean(datos_reales))
print("Media sesgada:", np.mean(datos_sesgados))
```

Ejercicio 2: Error de truncamiento

```python
import math
x = 1
aprox = 1 + x + (x**2)/2
real = math.exp(x)
error = abs(real - aprox)
print("aprox: ", aprox)
print("real: ", real)
print("error: ", error)
```

Ejercicio 1: Álgebra lineal con Numpy

```python
import numpy as np
A = np.array([[3,2], [1,2]])
b = np.array([5,5])
x = np.linalg.solve(A, b)
print(x)
```

Ejercicio 2: Optimización numérica con Scipy

```python
from scipy.optimize import minimize

f = lambda x: (x-2)**2
resultado = minimize(f, x0=0)
print(resultado.x)
```
