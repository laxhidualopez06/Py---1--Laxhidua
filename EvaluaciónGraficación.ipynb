```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set 1
cats = ['Norte', 'Sur', 'Este', 'Oeste']
conteos = np.array([120, 90, 150, 110])
# Una gráfica de barras es funcional ya que compara las entre las regiones los datos y se explica de una mejor manera
ax = plt.gca()
ax.bar(cats, conteos)

plt.xlabel('Regiones')
plt.ylabel('Conteo')
plt.title('Conteo por región')
ax.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
# Set 2.
n = 120
x = np.linspace(0, 100, n)
y = 2.0 * x + np.random.normal(0, 25, n)  # pendiente ~2 + ruido

#tamaño de la grafica
fig = plt.figure(figsize=(6,4))
ax = plt.gca()

# Paso 3) Dibujamos el scatter. No especificamos colores manualmente.
ax.scatter(x, y, alpha=0.75)

# Paso 4) Rotulamos y agregamos rejilla para lectura rápida.
ax.set_title('Dispersión o scatter')
ax.set_xlabel('Variable X')
ax.set_ylabel('Variable Y')
ax.grid(True, linestyle='--', alpha=0.4)

plt.show()
```

```python
# Set 3.
muestra = np.random.normal(loc=50, scale=10, size=1000) + np.random.exponential(scale=2, size=1000)*0.5
fig =plt.figure(figsize=(4,6))
ax = plt.gca()

#Histograma
ax.hist(muestra, bins=20, alpha=0.9)

#Lineas de media y mediana
media = np.mean(muestra)
mediana = np.median(muestra)
ax.axvline(media, linestyle='--', linewidth=1)
ax.axvline(mediana, linestyle='--', linewidth=1)
ax.annotate(f"Media - {media:.1f}", xy=(media, ax.get_ylim()[1]*0.9), xytext=(10,0), textcoords='offset points')
ax.annotate(f"Mediana - {mediana:.1f}", xy=(mediana, ax.get_ylim()[1]*0.8), xytext=(10,0), textcoords='offset points')

#Metadatos
ax.set_title("Grafica de Histograma")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True, axis='y', linestyle='--', alpha=0.4)

plt.show()

#Usé esta grafica para que los datos esten en un punto y cuales son los que estan separados
```

```python
# En el set 6 puse la grafica pie porque son pocos datos y son incomptatibles con otros datos

# Paso 1) Agregamos los valores y las categorias a la grafica Pie
categorias = ["A", "B", "C", "D"]
valores = np.array([35, 25, 22, 18])

# Paso 2) Agregamos el tamaño de la figura
fig = plt.figure(figsize = (6,6))
ax = plt.gca()

# Paso 3) Agregamos el nombre del titulo ademas creamos ya la grafica
ax.pie(valores, labels=categorias, autopct = "%1.1f%%", startangle=90)
ax.set_title("Circular: Participación por categoría")
ax.axis("equal")
ax.legend()

plt.show()
```

```python
# Set 7.
fs = 500
T  = 2.0
t = np.linspace(0, T, int(fs*T), endpoint=False)
f1, f2 = 50, 120
signal = 1.0*np.sin(2*np.pi*f1*t) + 0.6*np.sin(2*np.pi*f2*t) + 0.3*np.random.normal(0,1,len(t))

Y = np.fft.rfft(signal)
freqs = np.fft.rfftfreq(len(t), d=1/fs)
amplitud = np.abs(Y) / len(t)


fig = plt.figure(figsize=(6,4))
ax = plt.gca()
ax.plot(freqs, amplitud)
ax.set_xlim(0, 200)
ax.set_title('Análisis espectral: picos de frecuencia')
ax.set_xlabel('Frecuencia (Hz)')
ax.set_ylabel('Amplitud')
ax.grid(True, linestyle='--', alpha=0.4)

plt.show()
```

```python
# En el set 8 puse la grafica de burbujas porque esta tiene mas de dos variables

# Paso 1) Agregamos los datos de la grafica de burbujas
n = 60
X = np.random.uniform(10, 100, size = n)
Y = 1.5 * X + np.random.normal(0, 15, size = n)
S = np.random.uniform(50, 1500, size = n)

# Paso 2) agregamos la figura y el eje
fig = plt.figure(figsize = (6,4))
ax = plt.gca()

# Paso 3) Agregamos el scatter
ax.scatter(X, Y, s = S, alpha = 0.6)

# Paso 4) Agregamos la grafica de burbujas
ax.set_title("Burbujas: Tercera variable en tamaño")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid(True, linestyle= "--", alpha= 0.4)

plt.show()
```
