```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función
def f(x):
    return x**3 - 4*x - 9

# Intervalo dado
a = 2
b = 3

# Evaluación
fa = f(a)
fb = f(b)

print("f(a):", fa)
print("f(b):", fb)

# Verificación del Teorema del Valor Intermedio
if fa * fb < 0:
    print("Existe al menos una raíz en el intervalo")
else:
    print("No se garantiza raíz")

# Gráfica
x = np.linspace(2, 3, 100)
y = f(x)

plt.plot(x, y)
plt.axhline(0) # Línea del eje X para ver el cruce
plt.grid()
plt.show()
```

# PARTE B - Método de Bisección

```python
# a = 2 Limite inferior
# b = 3 Limite superior

# tolerancia = 1e ** -5

# iteraciones = 100
```

```python
import numpy as np

# Función y datos iniciales
def f(x):
    return x**3 - 4*x - 9

a = 2
b = 3
n = 5  # Número de iteraciones

# Ciclo de Bisección simple
for i in range(n):
    c = (a + b) / 2
    fc = f(c)

    # Lógica de bisección para actualizar el intervalo
    if f(a) * fc < 0:
        b = c  # La raíz está en el lado izquierdo
    else:
        a = c  # La raíz está en el lado derecho

    # Imprimir el avance de forma sencilla
    print("Iteración", i + 1, "-> Raíz aproximada:", c)

print("Resultado final después de 5 iteraciones:", c)
```

# PARTE C - Método de Aproximaciones Sucesivas

```python
import numpy as np

# Definición de la función iterativa g(x)
# Se obtuvo de x = raíz cúbica de (4x + 9)
def g(x):
    return (4*x + 9)**(1/3)

# Valores iniciales de tus apuntes
x = 2.5
tolerancia = 1e-5
max_iter = 100

# Ciclo de aproximaciones
for i in range(max_iter):
    x_nuevo = g(x)

    # Verificación de parada (error absoluto)
    if abs(x_nuevo - x) < tolerancia:
        break

    x = x_nuevo
    print("Iteración", i + 1, "-> x:", x)

print("Raíz aproximada por Punto Fijo:", x_nuevo)
print("Número de iteraciones:", i + 1)
```

# PARTE D -Interpolación

```python
import numpy as np

# Definición de la función original
def f(x):
    return x**3 - 4*x - 9

# Datos obtenidos (puntos cerca de la raíz)
x_datos = np.array([2.6, 2.7, 2.8])
y_datos = f(x_datos)

# Construcción del polinomio de grado 2
coeficientes = np.polyfit(x_datos, y_datos, 2)
polinomio = np.poly1d(coeficientes)

# Encontrar la raíz del polinomio (usando np.roots)
raices_polinomio = np.roots(coeficientes)

# Buscamos la raíz que esté en nuestro rango (cerca de 2.7)
raiz_interpolada = raices_polinomio[(raices_polinomio >= 2) & (raices_polinomio <= 3)]

print("Coeficientes del polinomio:", coeficientes)
print("Raíz encontrada mediante interpolación:", raiz_interpolada)

# Evaluación cerca de la raíz para comparar
x_eval = 2.706  # Valor cercano a la raíz real
print("Evaluación en x =", x_eval, "es:", polinomio(x_eval))
```
