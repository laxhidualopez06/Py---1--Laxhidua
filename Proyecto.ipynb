# **Limpieza de Datos**

```python
null_counts = data.isnull().sum()
columns_to_drop = null_counts.loc[null_counts > (0.03 * len(data))].index
data_cleaned = data.drop(columns=columns_to_drop)
data_cleaned.shape
```

```python
numeric_columns = data_cleaned.select_dtypes(include=['float64','int64']).columns
for col in numeric_columns:
    data_cleaned[col].fillna(data_cleaned[col].median(), inplace=True)

data_cleaned.isnull().sum()
```

```python
low_pop = data_cleaned[data_cleaned['track_popularity'] < data_cleaned['track_popularity'].quantile(0.25)]

# Agrupar por album_type y duración promedio
trend_data = low_pop.groupby(['album_type', 'track_duration_min']) \
                    .agg({'track_popularity': 'mean'}) \
                    .reset_index()

# Gráfica
plt.figure(figsize=(10,5))
plt.title('Relación entre Duración y Popularidad promedio por tipo de Álbum (tracks poco populares)',
          fontsize=14)

plt.xlabel('Duración del track (min)', fontsize=12)
plt.ylabel('Popularidad promedio', fontsize=12)

sns.lineplot(
    data=trend_data,
    x='track_duration_min',
    y='track_popularity',
    hue='album_type',
    marker='o'
)

plt.legend(title='Tipo de Álbum')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data['pop_category'] = pd.cut(
    data['track_popularity'],
    bins=[-1, 20, 50, 100],
    labels=['Baja', 'Media', 'Alta']
)

# Crear tabla dinámica para el heatmap
heatmap_data = data.pivot_table(
    values='track_duration_min',
    index='album_type',
    columns='pop_category',
    aggfunc='mean'
)

# Graficar
plt.figure(figsize=(8,5))
plt.title('Mapa de calor de DURACIÓN promedio según popularidad y tipo de álbum',
          fontsize=12)

sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.2f')

plt.xlabel('Categoría de Popularidad')
plt.ylabel('Tipo de Álbum')
plt.tight_layout()
plt.show()
```
