# **Ejercicio 1 - Regla del Trapecio** -- 16/04/2026

```python
import numpy as np # Importa la biblioteca NumPy para cálculos

def f(x): # Define la funcion a integrar
  return x**2 # Retorna el valor de la funcion x**2

a = 0 # Límite inferior de integración
b = 2 # Límite superior de integración
n = 100 # Número de subintervlos

h = (b-a)/n # Calcula el tamaño del paso

x = np.linspace(a, b, n+1) # Genera los puntos de discretización del intervalo

y = f(x) # Evalúa la función en todos los puntos

integral = (h/2) * (y[0] + 2 * np.sum(y[1:n]) + y[n]) # Aplica la fórmula del Trapecio compuesta

print("Integral aproximada (Trapecio):", integral) # Muestra el resultado numérico
```

# **Ejercicio 2 - Regla de Simpson** -- 16/04/2026

```python
a = 0 # Límite inferior
b = 2 # Límite superior
n = 100 # Número de subintervalos (debe ser par)

h = (b-a)/n # Calcula el tamaño del paso

x = np.linspace(a, b, n+1) # Genera los nodos del intervalo

y = f(x) # Evalúa la función

suma = y[0] + y[n] # Inicializa la suma con extremos

for i in range(1, n): # Itera sobre los puntos interiores
  if i % 2 == 0: # Verifica si el índice es par
    suma += 2*y[i]
  else:
    suma += 4*y[i]

integral = (h/3) * suma # Aplica la fórmula de Simpson

print("Integral aproximada (simpson): ", integral) # Muestra el resultado
```

# **Ejercicio 2 - Método de Trapecio en 2D**--> 20/04/2026

```python
import numpy as np

def f(x,y): # Define la función bidimensional
  return x**2 + y**2 # Función cuadrática

a, b = 0, 1 # Límites en x
c, d = 0, 1 # Límites en y

n = 50 # Número de puntos en x
m = 50 # Número de puntos en y

x = np.linspace(a,b,n) # Genera puntos en x
y = np.linspace(c,d,m) # Genera puntos en y

dx = (b-a)/(n-1) # Paso en x
dy = (d-c)/(m-1) # Paso en y

integral = 0 # Inicializa acumulador

for i in range(n): # Itera en x
  for j in range(m): # Itera en y

    peso = 1 # Inicializa peso

    if i==0 or i==n-1:
      peso *= 0.5 # Ajusta peso en borde
    if j == 0 or j == m-1: # Verifica si está en el borde en y
      peso *= 0.5 # Ajusta peso en borde

    integral += peso*f(x[i], y[j]) # Acumular valor ponderado

integral *= dx*dy # Multiplicar por el área del elemento

print("Integral (Trapecio 2D): ",integral) # Muestra resultado
```
